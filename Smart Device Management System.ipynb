```python
class SmartDevice:
    def __init__(self, name, device_id):
        self.name = name
        self.__device_id = device_id
        self.__power_status = "OFF"

    def get_device_id(self):
        return self.__device_id

    def get_power_status(self):
        return self.__power_status

    def turn_on(self):
        self.__power_status = "ON"
        print(self.name + " has been turned ON.")

    def turn_off(self):
        self.__power_status = "OFF"
        print(self.name + " has been turned OFF.")

    def display_info(self):
        print("\n--- Device Details ---")
        print("Name         :", self.name)
        print("Device ID    :", self.__device_id)
        print("Power Status :", self.__power_status)

class SecurityCamera(SmartDevice):
    def __init__(self, name, device_id):
        super().__init__(name, device_id)
        self.recording_status = False

    def start_recording(self):
        if self.get_power_status() == "OFF":
            print("Cannot record because the camera is OFF.")
            return
        self.recording_status = True
        print(self.name + " is now recording.")

    def stop_recording(self):
        self.recording_status = False
        print("Recording stopped on " + self.name)

    def display_info(self):
        super().display_info()
        if self.recording_status:
            print("Recording    : Yes")
        else:
            print("Recording    : No")

class SmartLight(SmartDevice):
    def __init__(self, name, device_id):
        super().__init__(name, device_id)
        self.brightness = 50

    def adjust_brightness(self):
        if self.get_power_status() == "OFF":
            print("Cannot adjust brightness because the light is OFF.")
            return

        level = int(input("Enter brightness level (0-100): "))
        if level >= 0 and level <= 100:
            self.brightness = level
            print("Brightness updated successfully.")
        else:
            print("Error: Value must be between 0 and 100.")

    def display_info(self):
        super().display_info()
        print("Brightness   : " + str(self.brightness) + "%")

class TemperatureSensor(SmartDevice):
    def __init__(self, name, device_id):
        super().__init__(name, device_id)
        self.temperature = 25.5

    def read_temperature(self):
        if self.get_power_status() == "OFF":
            print("Cannot read temperature because the sensor is OFF.")
            return
        print(self.name + " temperature reading is: " + str(self.temperature) + "°C")

    def display_info(self):
        super().display_info()
        print("Temperature  : " + str(self.temperature) + "°C")

camera = SecurityCamera("Front Gate Camera", "SC001")
light = SmartLight("Living Room Light", "SL001")
sensor = TemperatureSensor("Bedroom Sensor", "TS001")

while True:
    print("\n======== SMART DEVICE MANAGEMENT SYSTEM =========    ")
    print("1. Display Device Information")
    print("2. Turn Device On")
    print("3. Turn Device Off")
    print("4. Read Temperature")
    print("5. Adjust Brightness")
    print("6. Start/Stop Recording")
    print("7. Exit")

    choice = input("Enter your choice (1-7): ")

    if choice == "7":
        print("Exiting program. Goodbye!")
        break

    if choice == "1":
        print("\nWhich device info do you want to view?")
        print("1. Security Camera")
        print("2. Smart Light")
        print("3. Temperature Sensor")
        dev_choice = input("Select device: ")

        if dev_choice == "1":
            camera.display_info()
        elif dev_choice == "2":
            light.display_info()
        elif dev_choice == "3":
            sensor.display_info()
        else:
            print("Invalid choice.")

    elif choice == "2":
        print("\nWhich device do you want to turn ON?")
        print("1. Security Camera")
        print("2. Smart Light")
        print("3. Temperature Sensor")
        dev_choice = input("Select device: ")

        if dev_choice == "1":
            camera.turn_on()
        elif dev_choice == "2":
            light.turn_on()
        elif dev_choice == "3":
            sensor.turn_on()
        else:
            print("Invalid choice.")

    elif choice == "3":
        print("\nWhich device do you want to turn OFF?")
        print("1. Security Camera")
        print("2. Smart Light")
        print("3. Temperature Sensor")
        dev_choice = input("Select device: ")

        if dev_choice == "1":
            camera.turn_off()
        elif dev_choice == "2":
            light.turn_off()
        elif dev_choice == "3":
            sensor.turn_off()
        else:
            print("Invalid choice.")

    elif choice == "4":
        sensor.read_temperature()

    elif choice == "5":
        light.adjust_brightness()

    elif choice == "6":
        print("\nWhat action do you want to perform?")
        print("1. Start Recording")
        print("2. Stop Recording")
        action = input("Select action: ")
        if action == "1":
            camera.start_recording()
        elif action == "2":
            camera.stop_recording()
        else:
            print("Invalid action selection.")

    else:
        print("Invalid option. Please try again.")
```
